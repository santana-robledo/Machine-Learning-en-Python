## Funciones de Activación

<h2>Objetivo</h2><ul><li>Cómo aplicar diferentes funciones de activación en una Red Neuronal.</li></ul>


```python
import torch.nn as nn
import torch

import matplotlib.pyplot as plt
torch.manual_seed(2)
```

### Función Logística

Crea un tensor que vaya de -10 a 10:


```python
# Create a tensor

z = torch.arange(-10, 10, 0.1,).view(-1, 1)
```

Cuando usas *sequential*, puedes crear un objeto sigmoide:


```python
# Create a sigmoid object

sig = nn.Sigmoid()
```

Aplica la función sigmoide elemento a elemento con el objeto:


```python
# Make a prediction of sigmoid function

yhat = sig(z)
```

Grafica los resultados

```python
# Plot the result

plt.plot(z.detach().numpy(),yhat.detach().numpy())
plt.xlabel('z')
plt.ylabel('yhat')
```

Para módulos personalizados, llama a la sigmoide desde *torch* (nn.functional para la versión antigua), la cual aplica la sigmoide elemento a elemento desde el módulo de funciones y grafica los resultados:


```python
# Use the build in function to predict the result

yhat = torch.sigmoid(z)
plt.plot(z.numpy(), yhat.numpy())

plt.show()
```

### Tanh

Cuando usas *sequential*, puedes crear un objeto *tanh*:


```python
# Create a tanh object

TANH = nn.Tanh()
```

Llama el objeto y graficalo

```python
# Make the prediction using tanh object

yhat = TANH(z)
plt.plot(z.numpy(), yhat.numpy())
plt.show()
```

Para módulos personalizados, llama al objeto **Tanh** desde *torch* (nn.functional para la versión antigua), el cual aplica la función *tanh* elemento a elemento desde el módulo de funciones y grafica los resultados:


```python
# Make the prediction using the build-in tanh object

yhat = torch.tanh(z)
plt.plot(z.numpy(), yhat.numpy())
plt.show()
```

### Relu

Cuando usas *sequential*, puedes crear un objeto **ReLU**:


```python
# Create a relu object and make the prediction

RELU = nn.ReLU()
yhat = RELU(z)
plt.plot(z.numpy(), yhat.numpy())
```

Para módulos personalizados, llama al objeto **ReLU** desde nn.functional, el cual aplica la función elemento a elemento desde el módulo de funciones y grafica los resultados:


```python
# Use the build-in function to make the prediction

yhat = torch.relu(z)
plt.plot(z.numpy(), yhat.numpy())
plt.show()
```

## Compara Funciones de Activación

```python
# Plot the results to compare the activation functions

x = torch.arange(-2, 2, 0.1).view(-1, 1)
plt.plot(x.numpy(), torch.relu(x).numpy(), label='relu')
plt.plot(x.numpy(), torch.sigmoid(x).numpy(), label='sigmoid')
plt.plot(x.numpy(), torch.tanh(x).numpy(), label='tanh')
plt.legend()
```

Compara las funciones de activación con un tensor en el rango (-1, 1)


```python
x = torch.arange(-1, 1, 0.1).view(-1, 1)
plt.plot(x.numpy(), torch.relu(x).numpy(), label = 'relu')
plt.plot(x.numpy(), torch.sigmoid(x).numpy(), label = 'sigmoid')
plt.plot(x.numpy(), torch.tanh(x).numpy(), label = 'tanh')
plt.legend()
```
